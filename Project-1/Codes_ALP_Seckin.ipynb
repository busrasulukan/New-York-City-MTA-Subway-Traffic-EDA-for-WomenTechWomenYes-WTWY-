```python
mid_p_df["Date"] = pd.to_datetime(df['Date'], format='%m/%d/%Y',)
mid_p_df.dtypes
```

```python
mid_p_df['DAYS'] = mid_p_df['Date'].dt.day
mid_p_df['MONTH'] = mid_p_df['Date'].dt.month
mid_p_df['MONTHS'] =  mid_p_df['Date'].dt.strftime('%B')
mid_p_df['YEAR'] = mid_p_df['Date'].dt.year
mid_p_df['MONTHS/YEAR'] = mid_p_df['Date'].dt.strftime('%m/%Y')
mid_p_df['DAY/MONTHS/YEAR'] = mid_p_df['Date'].dt.strftime('%d/%m/%Y')
mid_p_df['DAY/MONTHS'] =  mid_p_df['Date'].dt.strftime('%d/%m')
mid_p_df['MONTH/DAY'] =  mid_p_df['Date'].dt.strftime('%m/%d')
mid_p_df['FULL DATE'] =  mid_p_df['Date'].dt.strftime('%Y/%m/%d')

mid_p_df.head()
```

```python
filtered_columns0 = mid_p_df[['sum_up_passengers', 'sum_up_all_ratios', 'DAYS' ,'DAY/MONTHS' ,'MONTHS','MONTHS/YEAR', 'YEAR', 'MONTH/DAY', 'FULL DATE']]
filtered_columns = mid_p_df[['sum_up_passengers', 'sum_up_all_ratios','YEAR', 'MONTH/DAY', 'FULL DATE']]
```

```python
DF2023= filtered_columns.loc[filtered_columns['YEAR'] == 2023]
DF2022= filtered_columns.loc[filtered_columns['YEAR'] == 2022]
DF2021= filtered_columns.loc[filtered_columns['YEAR'] == 2021]
DF2020= filtered_columns.loc[filtered_columns['YEAR'] == 2020]
```
